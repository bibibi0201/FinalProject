```python
import pandas as pd
from scipy.stats import ttest_rel

# อ่านไฟล์ Excel
df = pd.read_excel("pre_post_data.xlsx", sheet_name="pre-post")

# ดูข้อมูล .. แถวแรก
print(df.head(25))
```

```python
df_test = df.head(25)
df_test = df_test.apply(pd.to_numeric, errors='coerce')

print(df_test)
```

```python
# ทำ paired t-test
# Pain
t_stat_pain, p_val_pain = ttest_rel(df_test['Pain_Pre'], df_test['Pain_Post'])
print("\nPain t-test: t =", t_stat_pain, ", p =", p_val_pain)
```

```python
# Awareness
t_stat_awareness, p_val_awareness = ttest_rel(df_test['Awareness_Pre'],  df_test['Awareness_Post'])
print("Awareness t-test: t =", t_stat_awareness, ", p =", p_val_awareness)
```

```python
print("\nMean Pain Pre:", df_test['Pain_Pre'].mean())
print("Mean Pain Post:", df_test['Pain_Post'].mean())
print("Mean Awareness Pre:", df_test['Awareness_Pre'].mean())
print("Mean Awareness Post:", df_test['Awareness_Post'].mean())
```

## ผลสรุป Pre-Post Test

Pain (Pre-Post)
- **t = 0** → ค่าเฉลี่ยก่อน (Pain_Pre) กับหลัง (Pain_Post) **ไม่ต่างกันเลย**  
- **p = 1.0** → ไม่สามารถปฏิเสธสมมติฐานศูนย์ได้ (**ไม่มีความแตกต่างทางสถิติ**)  
- **สรุป:** การใช้อุปกรณ์หรือโปรแกรมนี้ยัง **ไม่ทำให้ระดับ Pain เปลี่ยนแปลง**

---

Awareness (Pre-Post)
- **t < 0** → ค่า Awareness_Post > Awareness_Pre  
  (เพราะเราใช้ paired t-test: post - pre)  
- **p < 0.05** → ผลต่างมี **นัยสำคัญทางสถิติ**  
- **สรุป:** หลังใช้อุปกรณ์ ผู้ใช้ **ตระหนักรู้ถึงท่านั่งของตัวเองมากขึ้นอย่างมีนัยสำคัญ**

- difference = post - pre

## สูตร Paired t-test:
$$
t = \frac{\text{mean difference}}{\text{std difference} / \sqrt{n}}
$$

## สูตรหา p-value (two-tailed) สำหรับ paired t-test

สำหรับ t-statistic ที่คำนวณจาก paired t-test:

$$
p = 2 \cdot \left( 1 - F_{t,\,n-1} \left( |t| \right) \right)
$$

โดยที่:  
- $t$ = t-statistic จาก paired t-test  
- $n$ = จำนวนคู่ข้อมูล  
- $F_{t,\,n-1}(x)$ = CDF ของ t-distribution กับ $df = n-1$  


```python
# สร้างตาราง mean
summary = pd.DataFrame({
    'Variable': ['Pain', 'Awareness'],
    'Pre Mean': [df_test['Pain_Pre'].mean(), df_test['Awareness_Pre'].mean()],
    'Post Mean': [df_test['Pain_Post'].mean(), df_test['Awareness_Post'].mean()]
})

# สร้างตาราง t-test
ttest = pd.DataFrame({
    'Variable': ['Pain', 'Awareness'],
    't-value': [t_stat_pain, t_stat_awareness],
    'p-value': [p_val_pain, p_val_awareness]
})

# รวมตารางเข้าด้วยกัน
final = pd.merge(summary, ttest, on='Variable')

# export เป็น Excel
final.to_excel('final_results.xlsx', index=False)

print("Export เสร็จแล้ว!")
```
